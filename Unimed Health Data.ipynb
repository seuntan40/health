### total billing amount by procedure

```python
# Calculate the total billing amount by summing Treatment Cost and Medication Cost
HD['Total Billing Amount'] = HD['Treatment Cost'] + HD['Medication Cost']

# Group by Procedure to get the sum of billing amounts
billing_by_procedure = HD.groupby('Procedure')['Total Billing Amount'].sum()

# Plotting the column chart
plt.figure(figsize=(10, 6))
billing_by_procedure.plot(kind='bar', color='blue')
plt.title('Total Billing Amount by Procedures')
plt.xlabel('Procedure')
plt.ylabel('Total Billing Amount')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

### Calculate Out of Pocket Cost as Total Billing Amount minus Insurance Coverage Amount

```python
# Calculate Out of Pocket Cost as Total Billing Amount minus Insurance Coverage Amount
HD['Out of Pocket Cost'] = HD['Total Billing Amount'] - HD['Insurance Coverage']

# Display the updated dataset with Out of Pocket Cost
print(HD[['Treatment Cost', 'Medication Cost', 'Total Billing Amount', 'Insurance Coverage', 'Out of Pocket Cost']].head())
```

#### Average Billing Amount 

```python
# Calculate the average billing amount
average_billing_amount = HD['Total Billing Amount'].mean()

# Display the average billing amount
print(f"The average billing amount is: ${average_billing_amount:.2f}")
```

### Lenght of stay 

```python
# Convert admission and discharge date columns to datetime if not already in datetime format
HD['Admitted Date'] = pd.to_datetime(HD['Admitted Date'])
HD['Discharge Date'] = pd.to_datetime(HD['Discharge Date'])

# Calculate Length of Stay in days
HD['Length of Stay'] = (HD['Discharge Date'] - HD['Admitted Date']).dt.days

# Display the updated dataset with Length of Stay
print(HD[['Admitted Date', 'Discharge Date', 'Length of Stay']].head())
```

### Average Billing per Procedure

```python
avg_cost_per_procedure = UNIMED.groupby('Procedure')['Total Billing Amount'].mean().reset_index()
avg_cost_per_procedure.columns = ['Procedure', 'Average Billing per Procedure']
UNIMED = UNIMED.merge(avg_cost_per_procedure, on='Procedure', how='left')
```

### Patient’s Age Group

```python
bins = [0, 18, 35, 50, 65, 100]
labels = ['0-18', '19-35', '36-50', '51-65', '65+']
HD['Age Group'] = pd.cut(HD['Age'], bins=bins, labels=labels)
print(HD[['Age', 'Age Group']].head())
```

### Total Cost per Department

```python
department_costs = HD.groupby('Department')['Total Billing Amount'].sum().reset_index()
department_costs.columns = ['Department', 'Total Cost per Department']
HD= HD.merge(department_costs, on='Department', how='left')

```

```python
HD['Readmission'] = HD.groupby('Patient ID')['Admitted Date'].diff().dt.days < 30
HD['Readmission'] = HD['Readmission'].fillna(False)
```

```python
HD
```

### Seasonal Admission Indicator

Useful for identifying seasonal trends in patient admissions or illnesses.

```python
HD['Season'] = HD['Admitted Date'].dt.month % 12 // 3 + 1
season_map = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'}
HD['Season'] = HD['Season'].map(season_map)
```

### MEDICATION COST PROPORTION
Identifies how much of the total cost is driven by medications, useful for pharmaceutical expense analysis.

```python
HD['Medication Cost Proportion'] = HD['Medication Cost'] / HD['Total Billing Amount']
```

```python
HD
```

#### Cost per Diagnosis Type

```python
cost_per_diagnosis = HD.groupby('Diagnosis')['Total Billing Amount'].mean().reset_index()
cost_per_diagnosis.columns = ['Diagnosis', 'Average Cost per Diagnosis']
HD = HD.merge(cost_per_diagnosis, on='Diagnosis', how='left')
```
